```python
import matplotlib
import matplotlib.pyplot as plt
```

# Histogram

```python
x = np.random.randn(200)
plt.style.use('classic')
plt.hist(x, bins=20, color='red')
plt.grid(True)
plt.xlabel('X-Axis')
plt.ylabel('Y-axis')
plt.title('A simple Histogram of Random numbers')
```

# Bar chart

```python
import pandas as pd
```

```python
data = {'Kids': ['Tayo', 'Akin', 'Bolu', 'Ada', 'Chisom'],
       'Age': [15,14,30,18,27],
         'Sex': ['M', 'M', 'F', 'F', 'F']}
Data = pd.DataFrame(data)
Data.head()
```

```python
plt.style.use('classic')
x = Data.Kids
y = Data.Age

plt.bar(x,y, color='blue', alpha= 0.7)
plt.xlabel('Sex')
plt.ylabel('Age')
```

# Pie chart

```python
subject = ['Mathematics', 'Chemistry', 'Physics', 'Biology']
students = [35,40,25,15]
colours = ['Red', 'Blue', 'Yellow', 'Green']
plt.pie(students, labels= subject, startangle = 90)
plt.show()
```

# Simple plot

```python
x = np.arange(1,10)
y = np.arange(11,20)
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y,marker='.',markersize=10,c ='black');
ax.set_title('X aagainst Y',fontsize=15,c='r')
ax.set_xlabel('X-axis',fontsize=16,c='r')
ax.set_ylabel('Y-axis',fontsize=16,c='r')
plt.xticks(fontsize=12);
plt.yticks(fontsize=12);
ax.grid(True)
plt.style.use('classic')
```
